# 【作业分析】

超市收银系统原型设计.已知超市有以下4种商品，信息如下表：
商品编号	商品名称	单价（元/斤）	库存（斤）
1	        苹果	    5.0            100
2	        香蕉	    3.0	           200
3           梨	        4.0	           150
4           桃子	    6.0	           50

### ☞【程序功能】
- 【1】 用户选择需要购买的商品编号并输入购买数量
- 【2】 系统计算并输出应付金额

### ☞【问题分析】
1. 求什么？（output）

2. 已知哪些有效数据才能够“求”？（input）

3. 从【已知】到【求】需要进行的数据处理过程（即：数学表示）？（prosess）

4. 【已知】如何获得？

5. 【数量】和【编号】能够满足应用需求的范围？

```c++
//框架代码
#include <stdio.h>
int main(void)
{
  int num;
  float weight,pay;

  //输入购买商品编号及数量
  scanf("%d%f",&num,&weight);
  if(num==1)
    pay=5.0*weight;
  if(num==2)
  	pay=3.0*weight;
  if(num==3)
    pay=4.0*weight;
  if(num==4)
    pay=6.0*weight;
  //输出应付金额
  printf("应付：%.2f元",pay);
}
```

    获取数据
    根据编码选项，进行核心数据计算：应付金额=单价*数量

    编码有效性判断
    数量有效性判断
